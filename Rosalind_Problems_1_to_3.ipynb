```python
with open('rosalind_revc.txt', 'r') as f:
    dna_str = f.read().strip()

# 2. 创建互补字典
complement_dict = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}

# 3. 先反转字符串 (Python 里的 [::-1] 表示倒序)
reversed_dna = dna_str[::-1]

# 4. 遍历反转后的字符串，一个一个查字典替换
# 这里用了一个列表推导式，看不懂没关系，先跑通
result_list = [complement_dict[base] for base in reversed_dna]
result = "".join(result_list)

print(result)
```
